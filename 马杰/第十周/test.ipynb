```python
import numpy as np

class NeuralNetwork:
    def __init__(self):
        # 初始化权重和偏置
        self.w1 = np.random.rand()
        self.w2 = np.random.rand()
        self.w3 = np.random.rand()
        self.w4 = np.random.rand()
        self.w5 = np.random.rand()
        self.w6 = np.random.rand()
        self.w7 = np.random.rand()
        self.w8 = np.random.rand()
        self.b1 = np.random.rand()
        self.b2 = np.random.rand()

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def sigmoid_derivative(self, x):
        return x * (1 - x)

    def forward(self, x1, x2):
        # 前向传播
        self.h1 = self.sigmoid(self.w1 * x1 + self.w2 * x2 + self.b1)
        self.h2 = self.sigmoid(self.w3 * x1 + self.w4 * x2 + self.b1)
        self.y1 = self.sigmoid(self.w5 * self.h1 + self.w6 * self.h2 + self.b2)
        self.y2 = self.sigmoid(self.w7 * self.h1 + self.w8 * self.h2 + self.b2)
        return self.y1, self.y2

    def loss(self, y1, y2, t1, t2):
        return 0.5 * ((t1 - y1)**2 + (t2 - y2)**2)

    def train(self, x1, x2, t1, t2, epochs, learning_rate):
        for epoch in range(epochs):
            # 前向传播
            y1, y2 = self.forward(x1, x2)

            # 计算损失
            current_loss = self.loss(y1, y2, t1, t2)

            # 反向传播
            dy1 = -(t1 - y1) * self.sigmoid_derivative(y1)
            dy2 = -(t2 - y2) * self.sigmoid_derivative(y2)

            dw5 = dy1 * self.h1
            dw6 = dy1 * self.h2
            dw7 = dy2 * self.h1
            dw8 = dy2 * self.h2
            
            dh1 = (dy1 * self.w5 + dy2 * self.w7) * self.sigmoid_derivative(self.h1)
            dh2 = (dy1 * self.w6 + dy2 * self.w8) * self.sigmoid_derivative(self.h2)
            
            dw1 = dh1 * x1  
            dw2 = dh1 * x2
            dw3 = dh2 * x1
            dw4 = dh2 * x2
            
            # 更新权重和偏置
            self.w1 -= learning_rate * dw1
            self.w2 -= learning_rate * dw2
            self.w3 -= learning_rate * dw3
            self.w4 -= learning_rate * dw4
            
            self.w5 -= learning_rate * dw5
            self.w6 -= learning_rate * dw6
            self.w7 -= learning_rate * dw7
            self.w8 -= learning_rate * dw8
            
            self.b1 -= learning_rate * (dh1 + dh2)
            self.b2 -= learning_rate * (dy1 + dy2)

            # 每1000轮打印一次损失
            if str(epoch)[-3:] == "999":
                print(f"轮次 {epoch+1}, 损失: {current_loss}")

# 创建神经网络实例
nn = NeuralNetwork()

# 训练数据
x1, x2 = 0.05, 0.10
t1, t2 = 0.01, 0.99

# 训练网络
nn.train(x1, x2, t1, t2, epochs=10000, learning_rate=0.1)

# 测试网络
y1, y2 = nn.forward(x1, x2)
print("输入:", x1, x2)
print("目标输出:", t1, t2)
print("实际输出:", y1, y2)
print("最终损失:", nn.loss(y1, y2, t1, t2))
```

keras

```python
from tensorflow.keras.datasets import mnist
from tensorflow.keras import models
from tensorflow.keras import layers
from tensorflow.keras.utils import to_categorical

# 加载数据集
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# 构建网络模型
network = models.Sequential()
network.add(layers.Dense(512, activation='relu', input_shape=(28*28,)))
network.add(layers.Dense(10, activation='softmax'))

# 编译模型
network.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])

# 数据预处理
train_images = train_images.reshape((60000, 28*28))
train_images = train_images.astype('float32') / 255

test_images = test_images.reshape((10000, 28*28))
test_images = test_images.astype('float32') / 255

# 标签转换为分类
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)

# 训练模型
network.fit(train_images, train_labels, epochs=5, batch_size=128)

# 评估模型
test_loss, test_acc = network.evaluate(test_images, test_labels, verbose=1)
print('test_loss: ', test_loss) 
print('test_acc: ', test_acc)

res = network.predict(test_images)

idx = 0
for i in range(res.shape[0]):
    for j in range(res.shape[1]):
        if res[i][j] == 1:
            print(f"第 {idx} 个图片的数字是 : {j}")
    idx += 1
```

```python
import numpy as np
import torch
import torch.nn as nn


class NeuralNetwork:
    def __init__(self,inputsize,hiddensize,outputsize,learningrate):
        self.inputsize = inputsize
        self.hiddensize = hiddensize
        self.outputsize = outputsize
        self.learningrate = learningrate

        self.w1 = torch.tensor(np.random.normal(0.0, pow(self.hiddensize, -0.5), (self.hiddensize, self.inputsize)), dtype=torch.float32)
        self.w2 = torch.tensor(np.random.normal(0.0, pow(self.outputsize, -0.5), (self.outputsize, self.hiddensize)), dtype=torch.float32)

        self.activation_function = nn.Sigmoid()
        
    def train(self,inputs_list,targets_list):
        inputs = torch.tensor(np.array(inputs_list,ndmin=2).T, dtype=torch.float32)
        targets = torch.tensor(np.array(targets_list,ndmin=2).T, dtype=torch.float32)

        hidden_inputs = torch.matmul(self.w1, inputs)
        hidden_outputs = self.activation_function(hidden_inputs)
        final_inputs = torch.matmul(self.w2, hidden_outputs)
        final_outputs = self.activation_function(final_inputs)

        output_errors = targets - final_outputs
        hidden_errors = torch.matmul(self.w2.T, output_errors)

        self.w2 += self.learningrate * torch.matmul((output_errors * final_outputs * (1 - final_outputs)), hidden_outputs.T).numpy()
        self.w1 += self.learningrate * torch.matmul((hidden_errors * hidden_outputs * (1 - hidden_outputs)), inputs.T).numpy()

    def query(self, inputs_list):
        inputs = torch.tensor(np.array(inputs_list, ndmin=2).T, dtype=torch.float32)
        hidden_inputs = torch.matmul(self.w1, inputs)
        hidden_outputs = self.activation_function(hidden_inputs)
        final_inputs = torch.matmul(self.w2, hidden_outputs)
        final_outputs = self.activation_function(final_inputs)
        return final_outputs


def main():
    inputsize = 28*28
    hiddensize = 100
    outputsize = 10
    learningrate = 0.1
    epochs = 5
    n = NeuralNetwork(inputsize,hiddensize,outputsize,learningrate)

    with open("dataset/mnist_train.csv", 'r') as train_data_file:
        train_data_list = train_data_file.readlines()

    with open("dataset/mnist_test.csv", 'r') as test_data_file:
        test_data_list = test_data_file.readlines()

    scores = []

    for i in range(epochs):
        for record in train_data_list:
            all_values = record.split(',')
            inputs = (np.asfarray(all_values[1:])) / 255.0 * 0.99 + 0.01
            targets = np.zeros(outputsize) + 0.01
            targets[int(all_values[0])] = 0.99
            n.train(inputs, targets)

    for test_img in test_data_list:
        all_values = test_img.split(',')
        correct_number = int(all_values[0])
        print("该图片对应的数字为:", correct_number)
        inputs = (np.asfarray(all_values[1:])) / 255.0 * 0.99 + 0.01
        outputs = n.query(inputs)
        label = np.argmax(outputs.detach().numpy())
        print("网络预测图片的数字是：", label)
        if label == correct_number:
            scores.append(1)
        else:
            scores.append(0)

    scores_array = np.asarray(scores)
    print("perfermance = ", scores_array.sum() / scores_array.size)
    

if __name__ == "__main__":
    main()
```
